#Расстояние Левенштейна

```python
def levenshtein_distance(s, t):
    if s == t:
        return 0
    elif len(s) == 0:
        return len(t)
    elif len(t) == 0:
        return len(s)

    rows = len(s) + 1
    cols = len(t) + 1
    distance = [[0 for _ in range(cols)] for _ in range(rows)]

    for i in range(1, rows):
        distance[i][0] = i
    for j in range(1, cols):
        distance[0][j] = j

    for j in range(1, cols):
        for i in range(1, rows):
            if s[i-1] == t[j-1]:
                cost = 0
            else:
                cost = 1
            distance[i][j] = min(distance[i-1][j] + 1,  # deletion
                                  distance[i][j-1] + 1,  # insertion
                                  distance[i-1][j-1] + cost)  # substitution

    return distance[rows-1][cols-1]

for i in range(0, len(vacancies)):
  print("\t\t для " , i , " вакансии")
  for j in range(0, len(courses)):
    print(levenshtein_distance(" ".join(list(vacancies[i])).split(), " ".join(list(courses[j])).split()))
```

#Коэффициент Жаккара

```python
def jaccard_similarity(tuple1, tuple2):
    set1 = set(tuple1)
    set2 = set(tuple2)

    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))

    jaccard = intersection / union if union != 0 else 0
    return jaccard

for i in range(0, len(vacancies)):
  print("\t\t для " , i , " вакансии")
  for j in range(0, len(courses)):
    print(jaccard_similarity(" ".join(list(vacancies[i])).split(), " ".join(list(courses[j])).split()))
```

#Евклидово расстояние

```python
import math

def euclidean_distance(vacancy_skills, course_skills):
    vacancy_skills_set = set(vacancy_skills)
    course_skills_set = set(course_skills)

    common_skills = vacancy_skills_set.intersection(course_skills_set)

    distance = math.sqrt(sum((vacancy_skills.count(skill) - course_skills.count(skill))**2 for skill in common_skills))

    return distance

for i in range(0, len(vacancies)):
  print("\t\t для " , i , " вакансии")
  for j in range(0, len(courses)):
    print(euclidean_distance(" ".join(list(vacancies[i])).split(), " ".join(list(courses[j])).split()))
```

```python
from gensim.models import KeyedVectors

# Загрузка предобученной модели word2vec
model_path = 'path_to_pretrained_model/word2vec.model'
word_vectors = KeyedVectors.load(model_path)

# Пример кортежей слов для тестирования
tuple1 = ('apple', 'orange', 'banana')
tuple2 = ('car', 'bike', 'bus')

# Вычисление word embeddings для каждого кортежа
embedding_tuple1 = sum(word_vectors[word] for word in tuple1) / len(tuple1)
embedding_tuple2 = sum(word_vectors[word] for word in tuple2) / len(tuple2)

# Вычисление косинусного расстояния между word embeddings кортежей
similarity = word_vectors.cosine_similarities(embedding_tuple1, [embedding_tuple2])[0]

print(f"Семантическая близость между {tuple1} и {tuple2} равна: {similarity}")
```
